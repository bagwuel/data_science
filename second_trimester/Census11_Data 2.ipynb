```python
def gender_age_count(data_frame):
    male_data = data_frame[data_frame['Gender'] == 'Male']
    female_data = data_frame[data_frame['Gender'] == 'Female']
    lower_band = data_frame['Age'].min()
    upper_band = lower_band + 4
    male_age_count = []
    female_age_count = []
    age_bands = []

    for i in range(data_frame['Age'].min(), data_frame['Age'].max(), 5):
        if data_frame['Age'].max() < lower_band + 10:
            upper_band = data_frame['Age'].max()
        age_bands.append(f'{lower_band} - {upper_band}')
        male_band_count = len(male_data.loc[(male_data['Age'] >= lower_band) & (male_data['Age'] <= upper_band)])
        female_band_count = len(female_data.loc[(female_data['Age'] >= lower_band) & (female_data['Age'] <= upper_band)])
        male_age_count.append(male_band_count)
        female_age_count.append(female_band_count)
        if upper_band == data_frame['Age'].max():
            break
        lower_band = lower_band + 5
        upper_band = upper_band + 5
    return([age_bands, female_age_count, male_age_count])

age_bands, female_age_count, male_age_count = gender_age_count(df1)
age_bands = age_bands[::-1]
female_age_count = female_age_count[::-1]
male_age_count = male_age_count[::-1]

sns.barplot(x = [-a for a in female_age_count], y = age_bands, color='green', label='Females')
sns.barplot(x = male_age_count, y = age_bands, color='gold', label='Males')

plt.xlabel('Count')
plt.ylabel('Age Bands')
plt.title('Age Pyramid by Gender')
plt.legend()

plt.show()
```

```python
child = df1[df1['Occupation'] == 'Child']
students = df1[df1['Occupation'].str.contains('Student')]
retired = df1[df1['Occupation'].str.contains('Retired')]
unemployed = df1[df1['Occupation'] == 'Unemployed']
employed = df1[~(df1['Occupation'] == 'Child') & ~(df1['Occupation'].str.contains('Student')) & ~(df1['Occupation'].str.contains('Retired')) & ~(df1['Occupation'] == 'Unemployed')]

x_axis = [child['Occupation'].count(), students['Occupation'].count(), retired['Occupation'].count(), unemployed['Occupation'].count(), employed['Occupation'].count()]
y_axis = ['Child', 'Student', 'Retired', 'Unemployed', 'Employed']

ax = sns.barplot(y=y_axis, x=[(a / sum(x_axis)) * 100 for a in x_axis])

plt.ylabel('Occupation')
plt.xlabel('Percentage (%)')
plt.title('Percentage Distribution of Occupations')

def annotate_bars(ax):
    for rect in ax.patches:
        y = rect.get_y() + rect.get_height() / 2
        x = rect.get_width()
        text = '{:.2f}%'.format(x)
        ax.text(x, y, text, ha='left' if x > 0 else 'right', va='center', fontsize=10, color='black')

annotate_bars(ax)

plt.show()
```

```python
#age distribution of religions with less than 100 members but greater than 3
data_frame = df1[df1['Religion'].isin(religion_under_100)]

muslim = data_frame[data_frame['Religion'] == 'Muslim']
sikh = data_frame[data_frame['Religion'] == 'Sikh']
jewish = data_frame[data_frame['Religion'] == 'Jewish']

lower_band = data_frame['Age'].min()
upper_band = lower_band + 4
muslim_age_count = []
sikh_age_count = []
jewish_age_count = []
age_bands = []

for i in range(data_frame['Age'].min(), data_frame['Age'].max(), 5):
    if data_frame['Age'].max() < lower_band + 10:
        upper_band = data_frame['Age'].max()
    age_bands.append(f'{lower_band} - {upper_band}')
    muslim_band_count = len(muslim.loc[(muslim['Age'] >= lower_band) & (muslim['Age'] <= upper_band)])
    sikh_band_count = len(sikh.loc[(sikh['Age'] >= lower_band) & (sikh['Age'] <= upper_band)])
    jewish_band_count = len(jewish.loc[(jewish['Age'] >= lower_band) & (jewish['Age'] <= upper_band)])
        
    muslim_age_count.append(muslim_band_count)
    sikh_age_count.append(sikh_band_count)
    jewish_age_count.append(jewish_band_count)
    if upper_band == data_frame['Age'].max():
        break
    lower_band = lower_band + 5
    upper_band = upper_band + 5

data = pd.DataFrame({
    'Age Bands': age_bands * 3,
    'Count': muslim_age_count + sikh_age_count + jewish_age_count,
    'Religion': ['Muslim'] * len(age_bands) + ['Sikh'] * len(age_bands) + ['Jewish'] * len(age_bands)
})

plt.figure(figsize=(10, 4.5))

sns.barplot(x='Age Bands', y='Count', hue='Religion', data=data)

plt.xticks(rotation=75)

plt.ylabel('Count', fontsize=16)
plt.xlabel('Age Bands', fontsize=16)
plt.title('Age distribution of Religions with members less than 100', fontsize=16)

plt.tick_params(labelsize=16)
[text.set_fontsize('16') for text in plt.legend().get_texts()]


plt.show()
```
